Learn visualization step by step

```
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```
# Load real dataset (restaurant data)
data = sns.load_dataset('tips')
data.head()
```

```
# Box Plot
# Shows distribution and outliers
sns.boxplot(x='day', y='total_bill', data=data)
plt.title('Box Plot - Daily Bills')
plt.show()
```

```
# Violin Plot
# Shows distribution shape
sns.violinplot(x='day', y='total_bill', data=data)
plt.title('Violin Plot - Bill Distribution')
plt.show()
```

```
# Heatmap
# Shows correlation between numeric columns
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```
# Pair Plot
# Shows relationships between all variables
sns.pairplot(data)
```
